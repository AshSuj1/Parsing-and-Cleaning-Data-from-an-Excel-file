## 3. Cleaning Data

#### The following steps will be performed in order to obtain a data frame with 202 country names, with 14 columns

* Drop columns with missing data
* Drop rows with missing data, and those rows that are not required
* Rename columns
* Set index to Country names
* Replace en-dash with an empty string

###### Note: The column names need not be extracted

#### 3.1 Drop Columns

```python
# Use Pandas drop function to remove unwanted columns

df = df.drop(['Unnamed: 0', 'Unnamed: 15', 'Unnamed: 17', 'Unnamed: 18', 'Unnamed: 19', 'Unnamed: 20',
       'Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23'], axis=1) 
df.columns
```

#### 3.2 Drop Rows

```python
# Use Pandas drop function to remove unwanted rows and also specify the index

# Drop rows after Zimbabwe
df = df.drop(df.index[206:245])

# Drop first few rows that hold missing data or unwated information
df = df.drop(df.index[0:4])
```

```python
# Display the new start of the altered data frame
df.head()
```

```python
# Display the new end of the altered data frame
df.tail()
```

#### 3.3 Rename columns

```python
# Uisng current column names as a pointer for where column names need to be altered

df = df.rename(columns={'TABLE 1. BASIC INDICATORS': 'Country names', 'Unnamed: 2': 0, 'Unnamed: 3': 1, 'Unnamed: 4': 2,
       'Unnamed: 5': 3, 'Unnamed: 6': 4, 'Unnamed: 7': 5, 'Unnamed: 8': 6, 'Unnamed: 9': 7,
       'Unnamed: 10': 8, 'Unnamed: 11': 9, 'Unnamed: 12': 10, 'Unnamed: 13': 11,
       'Unnamed: 14': 12, 'Unnamed: 16': 13})
```

```python
df.head()
```

#### 3.4 Set new index

```python
# Set index to country names and remove current index

df = df.set_index('Country names')
```

```python
# print df shape to confirm that the dimensins are (202, 14), i.e. 202 countries and 14 column names

df.shape
```

#### 3.5 Replace En dash

```python
# Replace en dash which signals missing data with an empt string

# Use a for loop to iterate over all the columns and use pandas replace function
# Use regex attribute to point to any non-alphanumeric character

for i in range(14):
    df[i].replace(regex=True,inplace=True,to_replace=r'\W',value=r'')
```

```python
df.head()
```

## 4. Save Data Frame to CSV file

```python
df.to_csv("basic_indicators.csv")
```
